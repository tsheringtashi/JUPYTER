```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
insurance_filepath = "../PYTHON/insurance.csv"
```

```python
insurance_df = pd.read_csv(insurance_filepath)
```

```python
sns.scatterplot(x=insurance_df['bmi'],y=insurance_df['expenses']);
```

```python
#The scatterplot above suggests that body mass index (BMI) and insurance charges are positively correlated, 
#where customers with higher BMI typically also tend to pay more in insurance costs. 
sns.regplot(x=insurance_df['bmi'],y=insurance_df['expenses']);
```

```python
sns.scatterplot(x=insurance_df['bmi'],y=insurance_df['expenses'],hue=insurance_df['smoker']);
```

```python
insurance_df.columns
```

```python
#notice that the regression line for smokers has a much steeper slope, relative to the line for nonsmokers
sns.lmplot(x='bmi',y='expenses',hue ='smoker',data = insurance_df);
```

```python
#on average, non-smokers are charged less than smokers, and
#the customers who pay the most are smokers; whereas the customers who pay the least are non-smokers.

sns.stripplot(x=insurance_df['smoker'],y=insurance_df['expenses']);
```
