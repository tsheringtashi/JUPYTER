```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('countries.csv')
```

```python
df_2007 = df[df['year'] == 2007]
```

```python
asia_2007 = df_2007[df_2007['continent']=='Asia']
europe_2007 = df_2007[df_2007['continent']=='Europe']
```

```python
plt.subplot(2,1,1)
plt.title('GDP Per Capita')
plt.hist(asia_2007['gdpPerCapita'],20, range=(0,50000),edgecolor = 'Black')
plt.ylabel('Asia')
plt.subplot(2,1,2)
plt.hist(europe_2007['gdpPerCapita'], 20, range=(0,50000), edgecolor = 'Black')
plt.ylabel('Europe');
```

### Compare Life Expectancy of Europe and America in 1997?

```python
df_1997 = df[df['year']== 1997]
```

```python
Europe_1997 = df_1997[df_1997['continent']=='Europe']
America_1997 = df_1997[df_1997['continent']=='Americas']
```

```python
df['continent'].unique()
```

```python
print(len(set(Europe_1997['country'])))
print(len(set(America_1997['country'])))
```

```python
print('mean lifeExpectancy of Europe')
print(Europe_1997['lifeExpectancy'].mean())
print('mean lifeExpectancy of America')
print(America_1997['lifeExpectancy'].mean())
print('median lifeExpectancy of Europe')
print(Europe_1997['lifeExpectancy'].median())
print('median lifeExpectancy of America')
print(America_1997['lifeExpectancy'].median())
```

```python
plt.subplot(2,1,1)
plt.title('LifeExpectancy in 1997')
plt.hist(Europe_1997['lifeExpectancy'],10, range = (55,85),edgecolor = 'Black')
plt.ylabel('Europe')
plt.subplot(2,1,2)
plt.hist(America_1997['lifeExpectancy'],10, range = (55,85), edgecolor = 'Black')
plt.ylabel('America');
```
