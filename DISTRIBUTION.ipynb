```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris_filepath ='../PYTHON/iris.csv'
```

```python
iris_data = pd.read_csv(iris_filepath)
```

```python
iris_data.head(2)
```

```python
sns.displot(iris_data['petal.length'],kde=False);
```

```python
sns.kdeplot(iris_data['petal.length'],shade =True);
```

```python
# Change the style of the figure to the "dark" theme
sns.set_style("dark")

# distribution chart 
plt.figure(figsize=(12,6))
sns.displot(iris_data['petal.length'],kde=False)
```

```python
#Seaborn has five different background themes: 
#(1)"darkgrid", (2)"whitegrid", (3)"dark", (4)"white", and (5)"ticks"
```

```python
# creating a 2D KDE plot with the sns.joinplot command.
sns.set_style('dark')
sns.jointplot(x = iris_data['petal.length'],y=iris_data['sepal.width'], kind='kde',shade=True);
```
